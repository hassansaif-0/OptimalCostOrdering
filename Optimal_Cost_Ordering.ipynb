```python

"""A CONTAINER CLASS FOR STORAGE OF OUR DATA"""
class Container:
    def __init__(self,pstring="",pcost=0.0,pdepth=0,pstat=""):
        self.pstring=pstring
        self.pcost=pcost
        self.pdepth=pdepth
        self.pstat=pstat
    def info(self):
        print("Permuted String With Minimum Cost is:",self.pstring)
        print("Minimum Cost String Value is:",self.pcost)
        print("Permuted String is found after  ",self.pstat,":",self.pdepth)
        



"""Cleanser Function For Data"""
datadict=dict()
glolist=list()
def cleanser(test):
    brac=False
    tem=""
    myl=list()
    for i in test:
        if(i==','and brac==True):
            tem=tem+i
        elif ((i==',' and brac==False)or i=='\n'):
            myl.append(tem)
            tem=""
        elif (i=='{'):
            tem=tem+i
            brac=True
        elif i=='}':
            tem=tem+i
            brac=False
        else:
            tem=tem+i
    return myl


"""Cost Calculating Function For Every Permutation"""
def cost(permuteval,inpdata):
    cost=0.0
    
    for i in range(len(permuteval)):
        clist=list()
        minval=99999.0
        valstr=""
        laterval=0.0
        signal=False;
        if(i==0):
            for x,y in inpdata.items():
                if x==int(permuteval[i]):
                    for key,val in y.items():
                        if(int(key)==x):
                            cost=cost+float(val)
                            signal=True
                            break
                if signal:
                    break
        
        else:
           
            for j in range(0,i):
                if(j>0):
                    clist.append(permuteval[j])
                    valstr=valstr+","+permuteval[j]
                    clist.append(valstr)
                else:
                    valstr=valstr+permuteval[j]
                    clist.append(valstr)
            
            for x,y in inpdata.items():
                checker=False
                if(x==int(permuteval[i])):
                    #print(x)
                    for key,val in y.items():
                        #print(key,val)
                        if key in clist:
                            if(float(val)<minval):
                                #print(float(val))
                                minval=float(val)
                                checker=True
                    if(not checker):
                        return -999
            #print(minval)
            cost=cost+minval
                        
                    
                            
        

                
                
    glolist.append(cost)       
    return cost
    




"""A STEP THAT FOR EACH TREE DEPTH SWAP CERTAIN VALUES TO FORM NEW  NODES"""
def swap(pval):
    
    retlist=list()
    v=len(pval)
    for i in range(1,v):
        sval=list(pval)
        sval[0],sval[i]=sval[i],sval[0]
        retlist.append(sval)
    return retlist
        
"""A UTILITY FUNCTION THAT CONVERTS AN INPUT LIST TO STRING"""   
def toString(List):
    return ''.join(List)            



"""A BFS APPROACH TO FORM NODES AT EACH LEVEL AND THEN FOR EACH NODE AT EACH LEVEL CALCULATE COST FUNCTION"""
def bfs(ival):
    retstr=""
    val=-99999
    fstr=""
    gcost=999999.0
    level=0
    Q=[]
    visited=[ival]
    
    Q.append(ival)
    while Q:
        vertex=Q.pop(0)
        retstr=(toString(vertex))
        val=cost(retstr,datadict)
        if(val<=gcost and val>=0):
            gcost=val
            fstr=retstr
            level+=1
        checker=swap(vertex)
        for i in checker:
            if i not in visited:
                visited.append(i)
                Q.append(i)
    obj=Container(fstr,gcost,level,"level")
    return obj




"""A DFS APPROACH TO FORM NODES AT EACH LEVEL AND THEN FOR EACH NODE AT EACH LEVEL CALCULATE COST FUNCTION"""
def dfs(ival):
    dretstr=""
    dval=-999999
    dfstr=""
    dgcost=999999.0
    depth=0
    Q=[]
    visited=[ival]
    
    Q.append(ival)
    while Q:
        vertex=Q.pop()
        dretstr=(toString(vertex))
        dval=cost(dretstr,datadict)
        if(dval<=dgcost and dval>=0):
            dgcost=dval
            dfstr=dretstr
            depth+=1
        checker=swap(vertex)
        for i in checker:
            if i not in visited:
                visited.append(i)
                Q.append(i)
    obj=Container(dfstr,dgcost,depth,"depth")
    return obj
               
            
"""UCS ALGORITHM FOR COST AND PATTERN EXPLORATION"""   
def UCS(ival):
    retstr=""
    val=-99999
    fstr=""
    gcost=999999.0
    level=0
    Q=[]
    visited=[ival]
    finallist=list()
    tlist=list()
    tlist.append(ival)
    retstr=(toString(ival))
    val=cost(retstr,datadict)
    tlist.append(val)
    Q.append(tlist)
    while Q:
        sorted(Q, key=lambda x:x[1])
        vertex=Q.pop(0)
        retstr=(toString(vertex[0]))
        val=cost(retstr,datadict)
        tlist=list()
        if(val<=gcost and val>=0):
            gcost=val
            fstr=retstr
            level+=1
        checker=swap(vertex[0])
        for i in checker:
            if i not in visited:
                visited.append(i)
                tlist.append(i)
                tlist.append(val)
                Q.append(tlist)
    obj=Container(fstr,gcost,level,"nodes")
    return obj

    
    
"""IMPLEMENTING FINAL FUNCTIONALITIES AND ALSO ADDING THE DRIVER CODE FOR OUR PROGRAMM"""
"""FILE NAME IS GIVEN HERE SAMPDATA YOU CAN REPLACE IT WITH ANY NAME AND PLACE THAT NAME FILE IN DIRECTORY"""

"""OPENING FILE FOR READING"""
f=open("sampdata.txt")

    
"""IGNORING THE LINE THAT CONTAIN NAME ROLL NUMBER AND ASSIGNMENT NUMBER AS PER REQUIREMENTS"""   
a=f.readline()
"""READING FIRST LINE"""

a=f.readline()
"""A TEST VARIABLE"""
uval=0
"""LOOPING TILL DATA IS READ"""
while(a):
    """PERFORMING CLEANSING ON EVERY LINE"""
    sval=""
    myl=cleanser(a)
    
    """CONVERTING FIRST VALUE TO INT"""
    myl[0]=int(myl[0])
    
    """IF NEW VALUE OCCUR MAKE A NEW KEY"""
    if(uval!=myl[0]):
        uval=myl[0]
        tdict=dict()
        datadict[uval]=tdict
        
    """STRIPPING OFF BRACES FOR PROCESSING"""
    myl[1]=myl[1].strip("{}")
    
    """IF NOT EMPTY STRING THEN FURTHER SPLLITING IT ON COMMAS"""
    if(len(myl[1])):
        test=myl[1].split(",")
        myl[1]=list(test)
        for i in range(len(myl[1])):
            myl[1][i]=int(myl[1][i])
        myl[2]=float(myl[2])
    else:
        myl[1]=str(myl[0])
        myl[2]=float(myl[2])
    """FOR INNER COST VALUES OF PARENT SET"""
    key=str(myl[1])
    for i in range(len(key)):
        if((key[i]=='[' or key[i]==']')or key[i]==' '):
            continue
        sval=sval+key[i]
    
    tdict[sval]=myl[2]    
        
    a=f.readline()
pstr=""
print("<<<<<<<<<<<------------------Displaying Data After Preprocessing------------>>>>>>>>>>>")
for x,y in datadict.items():
    print(x,y)
print("\n<<<-------------------------------------------------------------------------->>>\n")
"""FORMING A TEST COMBINATION INITILAYY BY COMBINING ALL KEYS"""
for x in datadict:
    pstr=pstr+str(x)

pval=list(pstr)

testobj=Container()
print("<<<<------Results From BFS------->>>>")
testobj=bfs(pval)
testobj.info()
print("\n\n<<<<------Results From DFS------->>>>\n\n")

testobj=dfs(pval)
testobj.info()
print("\n\n<<<<------Results From UCS------->>>>\n\n")

testobj=UCS(pval)
testobj.info()

```
